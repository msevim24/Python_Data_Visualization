### TASK-1: Recreate the Scatter Plot shown below

**The scatterplot attempts to show the relationship between the days employed versus the age of the person (DAYS_BIRTH) for people who were not unemployed. Note, to reproduce this chart you must remove unemployed people from the dataset first. Also note the sign of the axis, they are both transformed to be positive. Finally, feel free to adjust the *alpha* and *linewidth* parameters in the scatterplot since there are so many points stacked on top of each other.** 




##### Since the "DAYS_EMPLOYED" column holds the negative numbers and therefore shows the employed days, the numbers less than zero would mean the emplyed people. So, to exclude unemployed ones we can create a new DF.

```python
employed = df[df["DAYS_EMPLOYED"]<0]
```

##### Now, for this column, we can convert these negative numbers to positive ones.

```python
employed["DAYS_EMPLOYED"] = employed["DAYS_EMPLOYED"]*-1
```

```python
# Let's check the new column. Positive numbers...
employed["DAYS_EMPLOYED"]
```

##### We also need to change the "DAYS_BIRTH" column since it has negative values.

```python
employed["DAYS_BIRTH"] = employed["DAYS_BIRTH"]*-1
```

```python
# Let's check the change
employed["DAYS_BIRTH"]
```

```python
# Both columns as df
employed[["DAYS_BIRTH", "DAYS_EMPLOYED"]]
```

##### Now we can draw a scatterplot

```python
sns.scatterplot(employed, x="DAYS_BIRTH", y="DAYS_EMPLOYED");
```

##### Adjust the figure size, alpha and linewidth parameters.

```python
plt.figure(figsize=(10,6), dpi=100)
sns.scatterplot(employed, x="DAYS_BIRTH", y="DAYS_EMPLOYED", alpha=0.01, lw=1);

# alpha=0.01 --> to see the overlapping points better, low alpha value was chosen.
```

### TASK-4: Recreate the Heat Map shown below:



**This heatmap shows the correlation between the columns in the dataframe. You can get correlation with .corr() , also note that the FLAG_MOBIL column has NaN correlation with every other column, so you should drop it before calling .corr().**

##### Before drawing heatmaop let's see the correlation table.

Pandas dataframe.corr() is used to find the pairwise correlation of all columns in the Pandas Dataframe in Python. Any NaN values are automatically excluded.  To ignore any non-numeric values, use the parameter **numeric_only = True**.

**Syntax:** DataFrame.corr(self, method=’pearson’, min_periods=1, numeric_only = False) 

```python
# To see the only numeric columns... Otherwise, it gives error because of string values.
df.corr(numeric_only =True)
```

##### As it is seen from the table, there are lots of NAN values in "FLAG_MOBIL" column. We can drop it. Then we can draw the heatmap.

```python
sns.heatmap(df.drop('FLAG_MOBIL',axis=1).corr(numeric_only =True),cmap="viridis");
```

##### We can write the numbers on it using the annotation parameter.

```python
plt.figure(figsize=(10,5),dpi=100)
sns.heatmap(df.drop('FLAG_MOBIL',axis=1).corr(numeric_only =True),cmap="viridis", annot=True, lw=1);
```

```python
plt.figure(figsize=(10,5),dpi=100)
sns.heatmap(df.drop('FLAG_MOBIL',axis=1).corr(numeric_only =True),cmap="viridis", annot=True, lw=1);
```
