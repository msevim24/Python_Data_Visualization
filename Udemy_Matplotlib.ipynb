### Adding another set of axes to the Figure

```python
# Let's create two figures by plotting a and b and x and y
fig = plt.figure()

# Large Axes
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

# Small Axes
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])
axes2.plot(x,y)
```

```python
# Let's add title and axes names to the figure. This time I will plot a nad b two times

fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])                  # large figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25])     # small figure

# Set title, labels and limits for the large figure
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

# Set title, labels and limits for the small figure
axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")

axes1.plot(a,b)  # plot the first figure
axes2.plot(a,b)  # plot the second figure 
```

```python
# We can chance the place of the small figure
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])                
axes2 = fig.add_axes([1, 1, 0.25, 0.25])  # Change these parameters for small figure. (1,1) represents the right corner here.

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")

axes1.plot(a,b)  
axes2.plot(a,b)  
```

```python
# Make bigger the second figure
fig = plt.figure()

axes1 = fig.add_axes([0,0,1,1])                
axes2 = fig.add_axes([1, 1, 2, 2])  # Here, (2,2) represents width and hight of second figure

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel("A")
axes1.set_ylabel("B")
axes1.set_title("Power of 4")

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel("A")
axes2.set_ylabel("B")
axes2.set_title("Zoomed In")

axes1.plot(a,b)  
axes2.plot(a,b) 
```

```python
# Three figures

fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1]) # Full figure
axes2 = fig.add_axes([0.2, 0.5, 0.25, 0.25]) # Smaller figure
axes3 = fig.add_axes([1, 1, 0.25, 0.25]) # Starts at top right corner

# Larger Figure Axes 1
axes1.plot(a, b)

# Use set_ to add to the axes figure
axes1.set_xlabel('X Label')
axes1.set_ylabel('Y Label')
axes1.set_title('Big Figure')

# Insert Figure Axes 2
axes2.plot(a,b)
axes2.set_xlim(8,10)
axes2.set_ylim(4000,10000)
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoomed In');

# Insert Figure Axes 3
axes3.plot(a,b); # title, label and limit parameters were not used for the third figure
```

## Subplots

* **pyplot.subplots** creates a figure and a grid of subplots with a single call, while providing reasonable control over how the individual plots are created.
* The plt.subplots() call allows us to easily create Figure and Axes objects in side by side formations.
* The plt.subplots() command returns a tuple containing the Figure canvas and then a numpy array holding the axes objects.
* **(fig, axes):** "fig" is the entire Figure canvas and "axes" is a numpy array holding each of the axes according to position in the overall canvas.



* For more advanced use cases you can use **GridSpec** for a more general subplot layout or **Figure.add_subplot** for adding subplots at arbitrary locations within the figure.

#### plt.subplots()

The plt.subplots() object will act as a more automatic axis manager. This makes it much easier to show multiple plots side by side.

```python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes
fig, axes = plt.subplots()

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')  # color red
axes.set_xlabel('x')  # x label
axes.set_ylabel('y')  # y label
axes.set_title('title'); 
```

#### tight_layout()

A common issue with matplolib is overlapping subplots or figures. We ca use **fig.tight_layout()** or **plt.tight_layout()** method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

```python
# To deal with the spaces between graphs...
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(a,b)
axes[1][1].plot(a,b)
fig.tight_layout()
```

### Manual spacing on subplots: subplots_adjust()

Use **.subplots_adjust()** to adjust spacing manually.

For details: https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

For example ;
* left = 0.125  # the left side of the subplots of the figure
* right = 0.9   # the right side of the subplots of the figure
* bottom = 0.1  # the bottom of the subplots of the figure
* top = 0.9     # the top of the subplots of the figure
* wspace = 0.2  # the amount of width reserved for space between subplots,
              # expressed as a fraction of the average axis width
* hspace = 0.2  # the amount of height reserved for space between subplots,
              # expressed as a fraction of the average axis height

```python
# Another way to adjust figures
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][1].plot(x,y)
axes[1][0].plot(a,b)
axes[1][1].plot(a,b)
fig.subplots_adjust(wspace=0.6, hspace=0.3)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(12,8))  # we can use figsize as well

# Parameters at the axes level
axes[0][0].plot(a,b)
axes[1][1].plot(x,y)
axes[0][1].plot(y,x)
axes[1][0].plot(b,a)

# Use left,right,top, bottom to stretch subplots
# Use wspace,hspace to add spacing between subplots
fig.subplots_adjust(left=None,
    bottom=None,
    right=None,
    top=None,
    wspace=0.9,
    hspace=0.1,)
```

#### Set title, label and limit

```python
fig, axes = plt.subplots(nrows=2, ncols=2)
axes[0][0].plot(x,y)
axes[0][0].set_ylabel("Y LABEL 0,0")    # set ylabel

axes[0][1].plot(x,y)
axes[0][1].set_xlim(2,6)     # set xlimit

axes[1][0].plot(a,b)
axes[1][0].set_title("TITLE 1,0")    # set title
axes[1][1].plot(a,b)
axes[1][1].set_xlabel("X LABEL 1,1")     #set xlabel
fig.subplots_adjust(wspace=0.4, hspace=0.6)
```

## 3.Matplotlib Styling

### Legends

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, x**2, label="x**2")    # first plot with its legend
ax.plot(x, x**3, label="x**3")    # second plot with its legend
ax.legend()                       # we can specify its location inside the paranthesis
```

The legend function takes an optional keyword argument **loc** that can be used to specify where in the figure the legend is to be drawn. The allowed values of loc are numerical codes for the various places the legend can be drawn. Some of the most common loc values are:

```python
# Use "shift + tab to see the other options for "loc" parameter

ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location called "best"
fig
```

```python
# We can also assign a specific location

ax.legend(loc=(1.1,0.5)) # manually set location
fig

# 1.1 is the outside of the figure (on the right)
```

```python
ax.legend(loc=(-0.3,0.8)) # manually set location
fig

# "-0.3" is located on the left
```

### Setting colors, linewidths, linetypes

#### Colors with MatLab like syntax

We can define the colors of lines and other graphical elements in a number of ways. First of all, we can use the MATLAB-like syntax where "b" means blue, "g" means green, etc. The MATLAB API for selecting line styles are also supported: where, for example, "b.-" means a blue line with dots:

```python
# MATLAB style line color and style 
fig, ax = plt.subplots()
ax.plot(x, x**2, 'b.-') # blue line with dots
ax.plot(x, x**3, 'g--') # green dashed line
```

#### Colors with the color parameter

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**3, color="purple") 
```

```python
# We can use RGB code instead of the color name

fig, ax = plt.subplots()

ax.plot(x, x**2, color="blue", alpha=0.5) # half-transparant
ax.plot(x, x**3, color="#8B008B")        # RGB hex code (purple)
ax.plot(x, x**4, color="#FF8C00")        # RGB hex code (orange)
```

#### Linewidth

To change the line width, we can use the `linewidth` or `lw` keyword argument.

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use linewidth or lw
ax.plot(x, x-1, color="red", linewidth=0.25)
ax.plot(x, x-2, color="purple", lw=0.50)
ax.plot(x, x-3, color="blue", lw=2)
ax.plot(x, x-4, color="green", lw=10)
```

#### Linestyles

Possible linestype options: **‘--‘**, **‘–’**, **‘-.’**, **‘:’**, **‘steps’**...

```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x-1, color="red", lw=2, linestyle='-') # solid
ax.plot(x, x-2, color="green", lw=3, ls='-.')     # dash and dot
ax.plot(x, x-3, color="lightblue", lw=4, ls=':')      # dots
ax.plot(x, x-4, color="orange", lw=6, ls='--')     # dashes
```

#### Custom linestyle dash: set_dashes()

The dash sequence is a sequence of floats of even length describing the length of dashes and spaces in points.

For example, (5, 2, 1, 2) describes a sequence of 5 point dash, 2 point space, 1 point dash and 2 point space.

First, we see we can actually "grab" the line from the .plot() command

```python
fig, ax = plt.subplots(figsize=(12,6))

# custom dash
lines = ax.plot(x, x+8, color="black", lw=5)  # we assign it as "lines"
lines[0].set_dashes([10, 5]) # format: 10 point dash and 5 space
```

```python
fig, ax = plt.subplots(figsize=(12,6))
# custom dash
lines = ax.plot(x, x+8, color="orange", lw=5)
lines[0].set_dashes([1,1,1,1,10,10]) 

# one point, one space, 1 point, 1space, 10 point dash and 10 space, then repeat this order
```

#### Markers Style

Huge list of marker types can be found here: https://matplotlib.org/3.2.2/api/markers_api.html

Some of them:



```python
fig, ax = plt.subplots(figsize=(12,6))

ax.plot(x, x+3, color="lightblue", lw=2, marker="o", ls="--")  # circle marker
```

```python
fig, ax = plt.subplots(figsize=(12,6))

# Use marker for string code
# Use markersize or ms for size

ax.plot(x, x-1,marker='>',markersize=15)
ax.plot(x, x-2,marker='o',ms=15)       # ms can be used for markersize
ax.plot(x, x-3,marker='x',ms=15,lw=0)  # make linewidth zero to see only markers
ax.plot(x, x-4,marker='1',ms=15)
```

#### Custom marker edges, thickness, size, and style

```python
fig, ax = plt.subplots(figsize=(12,6))

# marker size and color
ax.plot(x, x, color="blue", lw=1, ls='-', marker='s', markersize=20, 
        markerfacecolor="red", markeredgewidth=4, markeredgecolor="orange");

# marker "s": square marker
# markersize 20: the size of these squares
# markerfacecolor: the color of the small square inside the big one (red)
# markeredgewidth: the difference between the both squares
# markeredgecolor: the color between the small and the big square (orange)
```
