```python
# Shading Regions with fill between() function

x = np.linspace(0, 10, 2000)
plt.figure(figsize= (10, 6))
plt.plot(x, np.sin(x), label = "Sin(x)")
plt.plot(x, np.cos(x), label = "Cos(x)")
plt.fill_between(x, 0, np.sin(x))
plt.fill_between(x, 0, np.cos(x))
plt.xlabel("X", fontsize = 18)
plt.ylabel("Y", fontsize = 18)
plt.title(" Sin(x)  &  Cos(x) ", fontsize= 14)
plt.legend(loc= "lower left") # legend will be placed at lower left position
plt.show()
```

```python
# # Display multiple plots in one figure using subplots()
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

# tanımlanan alan fig1. Bir alan belirliyoruz. Sonra oraya ax1'i çiziyoruz.
# baştan plt tanınmlandığı için x1 yazımlarında plt yazmıyoruz tekrar, set_ ile devam ediyoruz.

fig1, ax1 = plt.subplots(nrows = 3, ncols = 3, figsize= (20, 20)) # create a figure and subplots
ax1[0, 0].plot(x, y1, "tab:blue") # set the color of line chart
ax1[0, 0].set_title("Square Function") # setting title of subplot
ax1[0, 0].set_xlabel("X", fontsize = 18)
ax1[0, 0].set_ylabel("Y", fontsize = 18),
ax1[0, 0].set_facecolor("pink") 

ax1[0, 1].plot(x, y2, "tab:orange") # set the color of line chart
ax1[0, 1].set_title("Cubic Function") # setting title of subplot
ax1[0, 1].set_xlabel("X", fontsize = 18)
ax1[0, 1].set_ylabel("Y", fontsize = 18)

ax1[0, 2].plot(x, y3, "tab:green") # set the color of line chart
ax1[0, 2].set_title("Sine Function") # setting title of subplot
ax1[0, 2].set_xlabel("X", fontsize = 18)
ax1[0, 2].set_ylabel("Y", fontsize = 18)

ax1[1, 0].plot(x, y4, "r-") # set the color of line chart
ax1[1, 0].set_title("Cosine Function") # setting title of subplot
ax1[1, 0].set_xlabel("X", fontsize = 18)
ax1[1, 0].set_ylabel("Y", fontsize = 18)

ax1[1, 1].plot(x, y5, "b-") # set the color of line chart
ax1[1, 1].set_title("Tangent Function") # setting title of subplot
ax1[1, 1].set_xlabel("X", fontsize = 18)
ax1[1, 1].set_ylabel("Y", fontsize = 18)
ax1[1, 1].set_facecolor("yellow") 

ax1[1, 2].plot(x, y6, "g-") # set the color of line chart
ax1[1, 2].set_title("Hyperbolic Tangent Function") # setting title of subplot
ax1[1, 2].set_xlabel("X", fontsize = 18)
ax1[1, 2].set_ylabel("Y", fontsize = 18)

ax1[2, 0].plot(x, y7, "m-") # set the color of line chart
ax1[2, 0].set_title("Hyperbolic Sine Function") # setting title of subplot
ax1[2, 0].set_xlabel("X", fontsize = 18)
ax1[2, 0].set_ylabel("Y", fontsize = 18)

ax1[2, 1].plot(x, y8, "y*") # set the color of line chart
ax1[2, 1].set_title("Hyperbolic Cosine Function") # setting title of subplot
ax1[2, 1].set_xlabel("X", fontsize = 18)
ax1[2, 1].set_ylabel("Y", fontsize = 18)

ax1[2, 2].plot(x, y9, "kv") # set the color of line chart
ax1[2, 2].set_title("Exp Function") # setting title of subplot
ax1[2, 2].set_xlabel("X", fontsize = 18)
ax1[2, 2].set_ylabel("Y", fontsize = 18)
ax1[2, 2].set_facecolor("pink") 

plt.show()
```

# Group Bar Chart

```python
plt.figure(figsize = (7, 9))
# set width of bar
barWidth = 0.25
# set height of bar
y1 = np.array([17, 18, 29, 40])
y2 = np.array([20, 21, 22, 21])
y3 = np.array([5, 9, 11, 12])
# set position of bar on X axis
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]  # yana kaydırıyoruz 
pos3 = [x + barWidth for x in pos2]

# Make the plot
plt.bar(pos1, y1, color = "#FBC02D", width = barWidth, label = "Open")
plt.bar(pos2, y2, color = "#F57F17", width = barWidth, label = "Closed")
plt.bar(pos3, y3, color = "#E65100", width = barWidth, label = "Cancelled")

# Add xticks on the middle of group bars
plt.xlabel("Assignee", fontweight = "bold")
plt.ylabel("Number of Tickets", fontweight = "bold")
plt.xticks([i + barWidth for i in range(len(y1))], ["Isaac", "Asimov", "John", "Tolkein"])
# Soldaki yazılan değerler yerine sağdaki kategorik değerleri yazabiliriz. 4 grubu adlandır...
# Create legend & show graphic
plt.legend()
plt.show()
```

# Stacked Horizontal Bar

```python
# Grouped Bar Chart
plt.figure(figsize= (8, 5))
# set width of bar
barWidth = 0.25
# x1 daha önce isimler olarak tanımlanmıştı.Görsel horizontal olduğu için y ekseninde yer alıyor.
y1 = np.array([17, 18, 29, 40])
y2 = np.array([20, 21, 22, 21])
plt.barh(x1, y1, label = "Open Tickets", color = "#FF6F00")
plt.barh(x1, y2, label = "Closed Tickets",left = y1,  color = "#FFB300") # sol yani başlangıç y1 olacak. y2'yi üstüne...

plt.xlabel("X")
plt.ylabel("Y", rotation = 45)
plt.title("Bar  Chart")
plt.legend()
plt.show()
```

```python
plt.figure(figsize= (7, 9))

barWidth = 0.25

y1 = np.array([17, 18, 29, 40])
y2 = np.array([20, 21, 22, 21])

plt.barh(x1, y1, label = "Open Tickets", color = "#FF6F00",height=0.5) # horizonral olunca width değil hight
plt.barh(x1, y2, label = "Closed Tickets", left = y1, color = "#FFB300",height=0.5)

plt.xlabel("X")
plt.ylabel("Y", rotation = 45)

for x, y in zip(x1, y1):
    plt.text(y - 10 , x, "%d" % y, ha = "center", va = "bottom")  # y-10 ile içinde yazılacak yeri belirliyoruz
for x, y, z in zip(x1, y2, y1):  # burada z y1 değeri, aşağıdaki matematikte kullanıyoruz.
    plt.text(y + z - 10 , x, "%d" % y, ha = "center", va = "bottom") # y + z - 10 ile ilk değişkenin üzerine ekliyor.

plt.title("Bar  Chart")
plt.legend()
plt.show()

```

## Histogram

* Histogram, bir veri kümesinin frekans dağılımını görmemizi sağlar. Histogramlar, verilerin dağılımını göstermek için en sık kullanılan yöntemlerden biridir. 

```python
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = "yellowgreen")  #bins aralıkları/sütünları gösteriyor.
plt.gca().set(title = "Histogram", ylabel = "Frequency",xlabel="x")  #gca grafiği tanımlamak için kullanılan bir class
plt.show()
```

```python
# Using Edge Color for Readability
plt.figure(figsize = (10, 8))
x = np.random.normal(size=2000)
plt.hist(x, bins = 20, color = "orange", edgecolor = "black")   # edgecolor çerçeve için
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

```python
# Binnig
plt.figure(figsize = (5, 3))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 40, color = "orange", edgecolor = "black")
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
plt.figure(figsize = (5, 3))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 20, color = "blue", edgecolor = "black")
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
plt.figure(figsize = (5, 3))
x = np.random.normal(size = 2000)
plt.hist(x, bins = 10, color = "pink", edgecolor = "black")
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

```python
plt.figure(figsize = (8, 11))
x = np.random.normal(-4, 1, size = 800) # ortalama -4, standart sapması 1, 800 data
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)

# Plotting Multiple Histograms
plt.figure(figsize = (8, 11))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)
plt.hist(x, bins = 30, color = "yellowgreen", alpha = 0.6)
plt.hist(y, bins = 30, color = "blue", alpha = 0.6)
plt.hist(z, bins = 30, color = "pink", alpha = 0.6)
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

```python
plt.figure(figsize = (8, 11))
x = np.random.normal(-4, 1, size = 800)
y = np.random.normal(0, 1.5, size = 800)
z = np.random.normal(3.5, 1, size = 800)
plt.hist(x, bins = 30, color = "yellowgreen", alpha = 0.6)
plt.hist(y, bins = 30, color = "blue", alpha = 0.6)
plt.hist(z, bins = 30, color = "pink", alpha = 0.6)
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

```python
# Using Histogram to plot a cumulative distrubition function. cumulative=True ile önceki değerleri toplayarak çiziyor.
plt.figure(figsize = (5, 3))
x = np.random.rand(2000)
plt.hist(x, bins = 30, color = "orange", edgecolor = "black", cumulative = True)
plt.gca().set(title = "Histogram", ylabel = "Frequency")
plt.show()
```

# Area Plot

```python
#x = np.arange(1, 31)
#y = np.random.normal(10, 11, size = 30)
#y = np.square(y)
#plt.figure(figsize = (8, 4))
x = np.arange(1, 31)
y = np.random.normal(10, 100, size = 30)
y = np.square(y)
plt.figure(figsize = (8, 4))
plt.fill_between(x, y, alpha=0.2)
plt.plot(x, y, color = "blue") # changing fill color
plt.title(" Area  Chart", fontsize = 10)
plt.xlabel("X", fontsize = 10)
plt.ylabel("Y", fontsize = 10)
plt.show()
```

```python
x = np.arange(1, 31)
y = np.random.normal(10, 11, size = 30)
y = np.square(y)
plt.figure(figsize = (6, 4))
plt.fill_between(x, y, color = "#C8D700", alpha = 0.3)
plt.plot(x, y, color = "black")
plt.title(" Area  Chart", fontsize = 16)
plt.xlabel("X", fontsize = 16)
plt.ylabel("Y", fontsize = 16)
plt.show()
```
