#### Task One: Creating data from an equation

Your first task actually is pure numpy, then we will explore how to plot it out with Matplotlib. The [world famous equation](https://en.wikipedia.org/wiki/Mass%E2%80%93energy_equivalence) from Einstein:

$$E=mc^2$$


Use your knowledge of Numpy to create two arrays: E and m , where **m** is simply 11 evenly spaced values representing 0 grams to 10 grams. E should be the equivalent energy for the mass. You will need to figure out what to provide for **c** for the units m/s, a quick google search will easily give you the answer.

```python
m = np.linspace(0,10,11)
```

```python
m
```

```python
c = 3 * 10**8 # Speed of Light
```

```python
# the complete formula
E = m*c**2
```

```python
E
```

#### Task Two: Plotting E=mc^2

Can you plot this formula to see the relationship between Energy and Mass?

```python
plt.plot(m,E, "g--", linewidth=3)  # green dashed line and lw=3
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.title("E=mc^2");

# We could set limit for x or y as well, such as plt.xlim(0,10)
```

```python
# We can show it using figure object as well

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(m,E, color="lightblue", lw= 3);
```

#### Task Three

Can you figure out how to plot this on a logarthimic scale on the y axis? Place a grid along the y axis ticks as well.

```python
plt.plot(m,E, "g", lw=4)  # green line and linewidth=4
plt.xlabel("Mass in Grams")
plt.ylabel("Energy in Joules")
plt.title("E=mc^2")

plt.yscale("log")     # log scale on y
plt.grid(which="both", axis="y", color='b', alpha=0.5, linestyle='dashed', linewidth=0.5)

# "which=both" means major and minor axes
```

#### Task Four: Creating plots from data points

In finance, the yield curve is a curve showing several yields to maturity or interest rates across different contract lengths (2 month, 2 year, 20 year, etc. ...) for a similar debt contract. The curve shows the relation between the (level of the) interest rate (or cost of borrowing) and the time to maturity, known as the "term", of the debt for a given borrower in a given currency.

The U.S. dollar interest rates paid on U.S. Treasury securities for various maturities are closely watched by many traders, and are commonly plotted on a graph such as the one on the right, which is informally called "the yield curve".

We've obtained some yeild curve data for you from the [US Treasury Dept.](https://www.treasury.gov/resource-center/data-chart-center/interest-rates/pages/textview.aspx?data=yield). The data shows the interest paid for a US Treasury bond for a certain contract length. The labels list shows the corresponding contract length per index position.

Run the cell below to create the lists for plotting.

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

#### Task Five:  Figure out how to plot both curves on the same Figure. Add a legend to show which curve corresponds to a certain year.


```python
fig, ax = plt.subplots(figsize=(5,3))

ax.plot(labels, july16_2007, color="#FF8C00", alpha=0.5, label="july16_2007") # half-transparant (orange)
ax.plot(labels, july16_2020, color="#8B008B", label="july16_2020")        # RGB hex code (purple)
# do not forget labels, otherwise we cannot see the legend

ax.legend(loc="center right"); # legend

# We could use the code. The code of "center right" is "7"
# ax.legend(loc=7)

# Another option for legend
# ax.legend(loc=(0.7,0.5)); 
```

```python
# We could use figure object instead of subplots
fig = plt.figure(figsize=(5,3))

axes = fig.add_axes([0,0,1,1])  # left, bottom, width, height (range 0 to 1)
axes.plot(labels, july16_2007, label="july16_2007")
axes.plot(labels, july16_2020, label="july16_2020")

axes.legend(loc=7);
# plt.legend() # annother way to show legend
```

#### Task Six: The legend in the plot above looks a little strange in the middle of the curves. Place it ou *outside* the plot.

```python
fig, ax = plt.subplots()

ax.plot(labels, july16_2007, color="#FF8C00", alpha=0.5, label="july16_2007") 
ax.plot(labels, july16_2020, color="#8B008B", label="july16_2020")      

ax.legend(loc=(1.05,0.5)); # write a number more than 1 (on the y axis) to show it outsite the plot
```

```python
# Similarly, we could draw it using figure object

fig = plt.figure(figsize=(5,3))

axes = fig.add_axes([0,0,1,1])
axes.plot(labels, july16_2007, label="july16_2007")
axes.plot(labels, july16_2020, label="july16_2020")

plt.legend(loc=(1.05,0.5)); # this time I used "plt.legend()"
```

#### Task Seven: While the plot above clearly shows how rates fell from 2007 to 2020, putting these on the same plot makes it difficult to discern the rate differences within the same year. Use ".suplots()" to show each year's yield curve.

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(10,6), dpi=200)
axes[0].plot(labels, july16_2007)
axes[0].set_title("july16_2007")

axes[1].plot(labels, july16_2020)
axes[1].set_title("july16_2020")

# for enough space between two plots
fig.tight_layout()
```

#### Task Eight: Recreate the plot using twin axes.

```python
fig, ax1 = plt.subplots(figsize=(9,4), dpi=200) # add figsize and resolution (dpi)

ax1.plot(labels, july16_2007, lw=2, color="blue")
ax1.set_ylabel("2007", fontsize=12, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(labels, july16_2020, lw=2, color="red")
ax2.set_ylabel("2020", fontsize=12, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")


ax1.spines['left'].set_color('blue')  # blue line on the left
ax1.spines['left'].set_linewidth(3)

ax1.spines['right'].set_color('red')  # red line on the right
ax1.spines['right'].set_linewidth(3)

ax1.set_title("July 16th Yield Curves", fontsize=13)  # Title
```
