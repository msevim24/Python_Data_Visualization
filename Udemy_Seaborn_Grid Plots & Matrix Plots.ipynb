#  <span style ="color:#080f99;">1. Grid Plots
## Introduction
* Seaborn grid calls use Matplotlib subplots() to automatically create a grid based off a categorical column.
* Instead of passing in a specific number of columns or rows for the subplots, we can simply pass in the name of the column and seaborn will automatically map the subplots grid.
* Many of seaborn's built-in plot calls are running on top of this grid system. Directly calling the grid system allows users to heavily customize plots.

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# We need to upgrade Seaborn to see all annotations in heatmap, 
# otherwise annotations will appear only in the first row
!pip install seaborn --upgrade

import seaborn as sns
```

### Data

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

## 1.1. catplot()
We can choose categories, such as...



```python
# without specifying any "kind"
sns.catplot(data=df, x="gender", y="math score");
```

### kind

```python
sns.catplot(data=df, x="gender", y="math score", kind="box");
```

```python
# kind="violin" 

sns.catplot(data=df, x="gender", y="math score", kind="violin")
```

```python
# kind="bar"
sns.catplot(data=df, x="gender", y="math score", kind="bar")
```

### col and row
We can also choose some other columns as row or cols

```python
# Let's choose "lunch" feature as row. it will give us another dimension.

sns.catplot(data=df, x="gender", y="math score", kind="box", row="lunch");
```

```python
# Its display is different from "hue"
sns.catplot(data=df, x="gender", y="math score", kind="box", hue="lunch");
```

```python
# Let's try column

sns.catplot(data=df, x="gender", y="math score", kind="box", col="lunch");
```

```python
# col and row together; 2 rows and 2 columns. 
# By doing this, we can show two additional categorical variables

sns.catplot(data=df, x="gender", y="math score", kind="box", row="lunch", col="test preparation course");
```

```python
# Let's try violin plot. Change kind...

sns.catplot(data=df, x="gender", y="math score", kind="violin", col="lunch");
```

```python
# If there are too many categories, it will be hard to show and interpret.
sns.catplot(data=df, x="gender", y="math score", kind="violin", row="lunch", col="race/ethnicity");
```

## 1.2. PairGrid
Grid that pairplot is built on top of, allows for heavy customization of the pairplot.

```python
# Let's remember first the pairplot

sns.pairplot(data=df);
```

##### PairGrid gives us an option to customize pairplot according to different kind of plot on one pairplot. Using map_upper/lower/diag codes, we can draw different plots.

```python
# To show how it works, draw only scatterplot 

g = sns.PairGrid(df) # df is our data
g = g.map_upper(sns.scatterplot)  # display upper part of the plot as scatterplot 
```

```python
g = sns.PairGrid(df) # df is our data
g = g.map_upper(sns.scatterplot, color="red")    # upper--> scatter
g = g.map_diag(sns.histplot, color="green")                     # diagnal-->histplot
g= g.map_lower(sns.kdeplot);      # lower--> kdeplot
```

### hue
Additional variable

```python
g = sns.PairGrid(df, hue="gender") # df is our data
g = g.map_upper(sns.scatterplot)    # upper--> scatter
g = g.map_diag(sns.kdeplot, lw=2)   # diagonal--> kdeplot
g= g.map_lower(sns.kdeplot);        # lower--> kdeplot
```

### add_legend

```python
g = sns.PairGrid(df, hue="gender") # df is our data
g = g.map_upper(sns.scatterplot)    # upper--> scatter
g = g.map_diag(sns.kdeplot, lw=2)   # diagonal--> kdeplot
g= g.map_lower(sns.kdeplot);        # lower--> kdeplot

# Legend
g= g.add_legend()  # for hue parameter (male or female)
```

```python
g = sns.PairGrid(df, hue="gender", palette="viridis",hue_kws={"marker": ["o", "+"]})
g = g.map_upper(sns.scatterplot, linewidths=1, edgecolor="w", s=40)
g = g.map_diag(sns.distplot)
g = g.map_lower(sns.kdeplot)
g = g.add_legend();
```

## 1.3. FacetGrid

```python
sns.FacetGrid(data=df,col='gender',row='lunch')
```

```python
g = sns.FacetGrid(data=df,row='gender',col='lunch')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend();
```

### subplots_adjust
Adjust the space between plots

```python
g = sns.FacetGrid(data=df,col='gender',row='test preparation course')
g = g.map(plt.scatter, "math score", "reading score", edgecolor="w")
g.add_legend()

plt.subplots_adjust(hspace=0.3, wspace=2.5);
```

#  <span style ="color:#080f99;">2. Matrix Plots
## Introduction
* Matrix plots are the visual equivalent of displaying a pivot table.
* The matrix plot displays all the data passed in, **visualizing all the numeric values** in a DataFrame.
* **Note:** Not every DataFrame is a valid choice for a matrix plot such as a heatmap.
* There are two main matrix plot types:

  **1. heatmap():** Visually displays the distribution of cell values with a color mapping.
* Note that a heatmap should ideally have all cells be in the same units, so the color mapping makes sense across the entire DataFrame.
* In this particular case, all values were "rates" of percentage growth or change were in the heatmap.
  
  **2. clustermap():** Same visual as heatmap, but first conducts hierarchical clustering to reorganize data into groups.

### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data

The data below shows 2020 projections based on the World Population Prospects published by the UN.
For the estimates of the current year (2024), see https://www.ined.fr/en/everything_about_population/data/all-countries/?lst_continent=900&lst_pays=926

```python
df = pd.read_csv("country_table.csv")
df.head()
```

## 2.1. Heatmap
Syntax: 



```python
# It is helpful to set an index column. In this example, it will be "countries" column.

df = df.set_index("Countries")
df
```

```python
# Heatmap with entire dataset

sns.heatmap(df);
```

**Note:** "Life expectancy" column looks different because of its different unit. The other columns show rates but "Life expectancy" includes years.

```python
# We can drop life expectancy column and create a new df from rates columns

rates = df.drop("Life expectancy", axis=1)
```

```python
# Heatmap of rates
sns.heatmap(rates)
```

**Note:** After dropping "Life expectancy" column, notice that the correlation changed. Compare the heatmaps above. 

### linewidth
Make the boxes more visible

```python
sns.heatmap(rates, linewidth=0.5)
```

### annot=True
If True, write the data value in each cell. 

```python
plt.figure(dpi=200) # Make the immage sharper
sns.heatmap(rates, linewidth=0.5, annot=True);
```

### color mapping(cmap)

```python
plt.figure(figsize=(10,6), dpi=200) # Make the immage sharper
sns.heatmap(rates, linewidth=0.5, annot=True, cmap="viridis");
```

### center
The value at which to center the colormap when plotting divergent data. Using this parameter will change the default cmap if none is specified.

```python
plt.figure(figsize=(10,6), dpi=200) # Make the immage sharper
sns.heatmap(rates, linewidth=0.5, annot=True, cmap="viridis", center=40);  
```

```python
sns.heatmap(rates,linewidth=0.5,annot=True,cmap='viridis',center=5)
```

## 2.2. Clustermap
Similar to heatmap, but it also shows hierarchy.

Syntax:



```python
sns.clustermap(rates);
```

**Note:** From the clustermap above, we can see how effective the features on the countries, the similarity between countries/regions as well as the relationship between features.

### row/col_cluster
If True, cluster the rows or columns.

```python
# Remove column clusters
sns.clustermap(rates, col_cluster=False);
```

### cbar_pos
tuple of (left, bottom, width, height)

Position of the colorbar axes in the figure. Setting to None will disable the colorbar.

```python
sns.clustermap(rates,col_cluster=False,figsize=(12,8),cbar_pos=(-0.1, .2, .03, .4));
```

```python
# Let's use the similar parameters as we used for heatmap

plt.figure(figsize=(10,6), dpi=200) # Make the immage sharper
sns.clustermap(rates, linewidth=0.5, annot=True, cmap="viridis", col_cluster=False, center=40);  
```

#### Remove the index column name

```python
# Before using heatmap or clustermap, we can edit the DataFrame and drop the index name
rates.index.set_names('',inplace=True)
rates
```

Notice that there is no "Countries" index name, neither on the table above, nor in the clustermap below.

```python
sns.clustermap(rates,col_cluster=False,figsize=(12,8),cbar_pos=(-0.1, .2, .03, .4));
```
