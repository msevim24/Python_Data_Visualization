## Countplot

```python
sns.countplot(x="day", data=tips);  # hangi gün ne kadar yemek yendiği (kaç kişi)
```

```python
# burada bazılarını kaçırdım. Tekrar bak.
```

```python
## addition hue
fig, ax = plt.subplots(figsize = (12, 6))   # burada subplot yapmadan da ayn sonuç çıkardı.
ax = sns.countplot(x = "day", data = tips, hue = "smoker");
```

```python
fig, ax = plt.subplots(figsize = (12, 6))  
ax = sns.countplot(x = "day", data = tips, hue = "sex");
```

```python
fig, ax = plt.subplots(figsize = (12, 6))
ax = sns.countplot(x = "day", data = tips, hue = "size"); 
```

```python
# yüzdeleri bulabiliriz
# annotate grafiğe yorum yazdırmak, not almak için
# get_height column yüksekliklerine erişmek için
# tips.day.count: toplam yemek 244
# p.get_height() / tips.day.count(): örneğin 60/240=0.25
# p.get_x ile virgülden sonraki kısım herbirine git ve üzerine yaz anlamında konum belirtmek için

fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x = "day", data = tips);
for p in ax_tip.patches:   # patches grafige girip işlem yapmam için
    ax_tip.annotate(p.get_height() / tips.day.count(), (p.get_x(), p.get_height()));  
    # virgülden sonraki küsurat için aşağıdaki gibi sınırlama ve yuvarlama yapabiliriz.
```

```python
tips.day.count()  # verilen toplam yemek siparişi, satırlarımızın sayısı. 
```

```python
# round sayıları yuvarlamak için
# 2 ise virgülden sonraki küsuratı gösteriyor

fig, ax_tip = plt.subplots()        #ax_tip gibi bir variable'a eşitlersek daha hızlı çalışır.
ax_tip = sns.countplot(x = "day", data = tips);
for p in ax_tip.patches:   # patches grafige girip işlem yapmam için
    ax_tip.annotate(round(p.get_height() / tips.day.count(), 2), (p.get_x(), p.get_height()));     
```

```python
#üzerlerine grafikte y ekseninde karşılık gelen tam değerleri yazdıralım.
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x = "day", data = tips);
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x() + 0.3, p.get_height()));

# yükseliğini al, p içinde herbirine git ve p.get'in tepe noktasına annotate yap.
# p.get_x, x eksenine göre; p.get_height, y eksenine göre (sağa sola, aşağı yukarı)
```

```python
# bu sefer de direkt % işaretiyle yazdıralım.
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x = "day", data = tips);
for p in ax_tip.patches:
    ax_tip.annotate((p.get_height()), (p.get_x() + 0.3, p.get_height()));
    ax_tip.annotate("%" + str(round(p.get_height()/tips.day.count(), 2)*100), (p.get_x()+0.2,p.get_height()-15))
# 0.3 yazıyı sağa sola, -15 ise gethight'ın altına üstüne alıyor. Burada altına aldı.
```

## Bar Plot

```python
# kadın ve erkek bahşiş oranları.
sns.barplot("sex", "tip", data=tips);  
# x ve y yazmadan yazdık.Dik çizgiler varyans
# matplot barplota varyans eklemezken seaborn ekliyor.
```

```python
# sum estimator
# kadınlar ve erkekler toplamda ne kadar tip vermiş.
sns.barplot("sex", "tip", data=tips, estimator=np.sum);
```

```python
# standart sapmalarını da görebiliriz. ci metodunun içine sd yazarak
sns.barplot("sex", "tip", data=tips, ci="sd");  
```

```python
# üstlü altlı değil de yan yana hem y eksenindeki sayıyı hem de yüzdesini yazdırabiliriz.
fig, ax_tip = plt.subplots()
ax_tip = sns.countplot(x = "day", data = tips);
for p in ax_tip.patches:
    ax_tip.annotate("(" +
                   str(p.get_height()) +
                   ") " +
                   "%" +
                   str(round((p.get_height()) / tips.day.count(), 2) * 100), (p.get_x() + 0.07, p.get_height() -10));
```

## Bar and Count Plots

Count plot bir değişkenden kaç tane olduğunu sayar, hangi gün ne kadar yemek yenmiş. Barplot'ta ise iki değişkene, mesela günlere göre total bill'e bakabiliyoruz. Ayrıca varyansları da gösteriyor.

```python
fig, ax=plt.subplots(nrows=1, ncols=2, figsize=(16,6)) # buradaki nrows ve ncols default değerler yazmamıza gerek yok
ax[0]=sns.barplot(data=tips, x="day", y="total_bill", ax=ax[0]) # ax[0] ve ax[1]'leri yazmasak da olur.
ax[1]=sns.countplot(data=tips, x="day", ax=ax[1]);
```

```python
fig, ax=plt.subplots(1, 2, figsize=(16,6)) # buradaki nrows ve ncols default değerler yazmamıza gerek yok
sns.barplot(data=tips, x="day", y="total_bill", ax=ax[0])
sns.countplot(data=tips, x="day", ax=ax[1]);
```

```python
# her yılı aylara göre gruplayabiliriz. groupby()
flights.groupby(["year", "month"]).sum()
```

```python
#daha detaylı inişli ve çıkışlı zamanserisi grafiğini görebiliriz.
y=flights.groupby(["year", "month"]).sum().values
x=flights.index
sns.lineplot(x=flights.index, y=flights.passengers.values);
```

```python
#yılları saturlara ayları columnlara 
flights_wide = flights.pivot("year", "month", "passengers")
flights_wide.head()
```

```python
# yukarıdaki değişkeni kullanarak her yılı ayrı ayrı grafikle görebiliriz.
plt.figure(figsize=(12,6))
sns.lineplot(data=flights_wide.T);   # Transpose aldık.
```

```python
#pivot tabloya girmeden şöyle de yazdırabilirdik.
plt.figure(figsize=(12,6))
sns.lineplot(data=flights, x="month", y="passengers", hue="year", palette="viridis")
plt.legend(loc=(1.1, .5))    # loc x ve y eksenindeki pozisyonu gösteriyor.
plt.show()

```

```python
# pointplot olarak da yazdırabiliriz
plt.figure(figsize = (12, 6))
sns.pointplot(data = flights, x = "month", y = "passengers", hue = "year", palette= "viridis")
plt.legend(loc = (1.1, .5))
plt.show()
```
