## Introduction
* Comparison plots are essentially 2-dimensional versions of the plots.
* There are two main plot types discussed here: **jointplot()** and **pairplot()**.
#### 1. jointplot():
* We can map hidtograms to each feature of a scatterplot to clarify the distributions within each feature.
* We can also adjust the scatterplot to be a hex plot or a 2D KDE plot.
* Hexagons show the density of the variation with dark or light points.



*2D KDE plots show shaded distribution between both KDEs:



#### 2. pairplot():
* It is a quick way to compare all numerical columns in a DataFrame.
* It automatically creates a histogram for each column and a scatterplot comparison between all possible combinations of columns.
* **Warning:** pairplot() can be CPU and RAM intensive for large DataFrames with many columns. It is a good idea to first filter down to only the columns you are interested in.



### Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### Data

```python
df = pd.read_csv("StudentsPerformance.csv")
df.head()
```

## 1. jointplot()

```python
# Let's discover the relationship between math and reading scores

sns.jointplot(data=df, x="reading score", y="math score");
```

```python
# We can use alpha to look at the points closely

sns.jointplot(data=df, x="reading score", y="math score", alpha=0.2);
```

#### kind

```python
# By default its "kind" is scatter, but we have other options

sns.jointplot(data=df, x="reading score", y="math score", kind="hex"); # hexagon
```

```python
# let's try "hist"
sns.jointplot(data=df, x="reading score", y="math score", kind="hist");
```

```python
# Another kind is kde

sns.jointplot(data=df, x="reading score", y="math score", kind="kde");
```

```python
# We can shade this kde version

sns.jointplot(data=df, x="reading score", y="math score", kind="kde", shade=True);
```

```python
# Let^s add gender as hue to the default scatter kind

sns.jointplot(data=df, x="reading score", y="math score", hue="gender");

# because of this hue, it will create two separate kde plots and colorful points according to variable in hue.
```

## 2. pairplot()
It is recommended to use it for selected features instead of all columns.

```python
# for all columns

sns.pairplot(data=df);
```

```python
# Add hue

sns.pairplot(data=df, hue="gender");
```

```python
# Add palatte

sns.pairplot(df,hue='gender',palette='viridis')
```

#### diag_kind

```python
# We can choose the kind as we did above

sns.pairplot(data=df, hue="gender", diag_kind="hist", palette="Set1");
```

#### corner=True

To remove duplicates

```python
sns.pairplot(data=df, hue="gender", corner=True, palette="rocket");
```
