```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
# data sets in seaborn
print(sns.get_dataset_names())
```

```python
# Load the data
tips = sns.load_dataset("tips")
```

```python
tips.sample(10)
```

```python
tips.info()
```

```python
tips["size"].unique() # unique values in "size"
```

### Scatter Plot

```python
sns.scatterplot(data = tips, x = "total_bill", y = "tip")
plt.show()
```

```python
# change the size of the figure
plt.figure(figsize= (10, 5))
sns.scatterplot(x = tips.total_bill, y = tips.tip)
plt.show()
```

```python
# add a third variable with hue (smoker or not)
plt.figure(figsize= (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker")  # hue must be categorical
plt.show()
```

```python
plt.figure(figsize= (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "sex")  # hue must be categorical
plt.show()
```

```python
plt.figure(figsize= (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "day")  # hue must be categorical
plt.show()
```

```python
# another variable with style as sex (male/female)
# by using style we can see categories in differend shapes (+,x,., etc)
plt.figure(figsize= (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex") 
plt.show()
```

```python
# let's look at total_bill and tip according to days
plt.figure(figsize= (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "day")  
plt.show()
```

```python
# using style and hue together we can have a colorful shapes showing categories 
plt.figure(figsize= (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "day", hue = "day")  
plt.show()
```

```python
# 4 variables and many categories in a graph (tip(y), total_bill, day, sex)
plt.figure(figsize= (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "day", hue = "sex") 
plt.show()
```

```python
plt.figure(figsize= (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "smoker", hue = "sex") 
plt.show()
```

```python
# draw several graphs by using fig and hg
fig, hg = plt.subplots(1, 2, figsize = (14, 8))  # 1 row, 2 columns
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex", ax = hg[0]) # ax shows the order of the graph
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", ax = hg[1])
plt.show()
```

```python
fig, hg = plt.subplots(2, 1, figsize = (14, 8)) # 2 rows, 1 column
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex", ax = hg[0])
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", ax = hg[1])
plt.show()
```

```python
# us "palette" to change colors
plt.figure(figsize= (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", hue = "smoker", palette= "plasma")
plt.show()
```

```python
# use "alpha" for opacity; default value is 1
plt.figure(figsize= (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", alpha= 0.3555555)
plt.show()
```

```python
plt.figure(figsize = (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex", hue = "day", alpha = 0.7, palette= "plasma");
```

```python
# with size(s), make a bigger shapes
plt.figure(figsize = (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex", hue = "day", alpha = 0.7, s = 200);
```

```python
# palette= "cubehelix"
plt.figure(figsize = (10, 8))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", style = "sex", hue = "day", alpha = 0.9, s = 500, palette= "cubehelix");
```

## Distribution Plot

```python
sns.displot(x = "total_bill", data = tips);
# the highest one shows mode
```

```python
# adjust the number of bins 
sns.displot(x = "total_bill", data = tips, bins = 50);
```

```python
sns.displot(x = "total_bill", data = tips, bins = 5);
```

```python
# line plot with kernel density estimation (kde)
sns.displot(x = "total_bill", data = tips, bins = 50, kde = True);
```

```python
sns.displot(x = "total_bill", data = tips, bins = 5, kde = True); # kde = kernel densty estimation
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 8))

sns.scatterplot(data = tips, x = "total_bill", y = "tip", s = 200, hue = "sex", ax = ax[0])

sns.displot(data = tips, x = "total_bill", ax = ax[1]); # use hisplot, it would be better
```

```python
fig, hg = plt.subplots(1, 2, figsize = (10, 5))
sns.scatterplot(data = tips, x = "total_bill", y = "tip", s = 300, hue = "sex", ax = hg[1])
sns.histplot(data = tips, x = "tip", ax = hg[0]);   # ax = axis   
```

```python
# show hist and kde as different graphs
sns.displot(data = tips, x = "total_bill", kind = "hist");
sns.displot(data = tips, x = "total_bill", kind = "kde");
```

```python
sns.displot(data = tips, x = "total_bill", kind = "hist", kde = True);
```

```python
# use "aspect" to adjust width; defalut value is 1
sns.displot(data = tips, x = "total_bill", kind = "hist", kde = True, aspect = 5)
plt.show()
```

### Hist plot

```python
plt.figure(figsize= (10, 5))
sns.histplot(data = tips, x = "total_bill", bins = 50);
```

```python
# rugplot
fig, ax = plt.subplots(1, 2, figsize = (10, 5))
sns.rugplot(data = tips, x = "total_bill", height= 0.8, ax = ax[0]);
sns.histplot(data = tips, x = "tip", kde = True, ax = ax[1]);
```

```python
# use "style" to change background
sns.set(style = "whitegrid")
sns.histplot(data = tips, x = "tip", kde = True);
```

```python
sns.set(style = "dark")
sns.histplot(data = tips, x = "tip", kde = True);
```

### KDE Plot (kernel density estimation)

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.sample(10)
```

```python
# see unique values of species
penguins.species.unique()
```

```python
# see unique values of island
penguins.island.unique()
```

```python
penguins.info()
```

```python
sns.kdeplot(data = penguins, x = "body_mass_g");
```

```python
sns.set(style = "darkgrid")
```

```python
sns.kdeplot(data = penguins, x = "body_mass_g");
```

```python
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "sex");
```

```python
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species");
```

```python
# use "shade" to fill the background of density plots
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "species", shade = True);
```

```python
sns.kdeplot(data = penguins, x = "body_mass_g", hue = "sex", shade = True);
```

```python
# 2D kde graph (for tips data)
sns.kdeplot(data = tips, x = "total_bill", y = "tip"); 
```

```python
sns.kdeplot(data = tips, x = "total_bill", y = "tip", color = "red");
```

```python
sns.kdeplot(data = tips, x = "total_bill", y = "tip", color = "red", shade = True);
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 5))

sns.kdeplot(data = tips, x = "total_bill", y = "tip", color = "red", shade = True, ax = ax[0]);
sns.scatterplot(data = tips, x = "total_bill", y = "tip", color = "red", ax = ax[1]);
#see outliers with scatterplot
```

```python
fig, ax = plt.subplots(1, 2, figsize = (15, 5))

sns.kdeplot(data = tips, x = "total_bill", y = "tip", color = "blue", shade = True, ax = ax[1]);
sns.scatterplot(data = tips, x = "total_bill", y = "tip", color = "red", ax = ax[0]);
```

```python
# draw rugplot, kde and histplot together
sns.displot(data = penguins,
           x = "body_mass_g",
           bins = 10,
           rug = True,
           kde = True);
```

## Jointplot

```python
# with jointplot, we can see vertical and horizontal density in a graph
plt.figure(figsize = (10, 15))
sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "scatter");
```

```python
# use kind to change main graph. Here it is hex
plt.figure(figsize = (10, 15))
sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "hex");
```

```python
plt.figure(figsize = (10, 15))
sns.jointplot(data = tips, x = "total_bill", y = "tip", kind = "kde", shade = True, color = "green");
plt.show()
```

### Pairplot

```python
# to draw all numeric variables, use pairplot
# we can see distribution plots and scatter plots
sns.pairplot(data = tips);
```
